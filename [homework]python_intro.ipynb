### Почти двойной факториал
Дано натуральное $N \leqslant 100$.  Выведите произведение всех нечётных натуральных чисел, не превосходящих $N$.

```python
N = 25

def almost_double_factorial(n):
  a = 1
  if n == 0:
    return a
  else:
    for i in range(1,n+1):
      if i%2 == 1:
        a *= i
    return a

N_double_fact = almost_double_factorial(N)

assert N_double_fact == 7905853580625, "Something is wrong! Please try again"

# def almost_double_factorial(n):
#     res = 1
#     for i in range(1, n+1, 2):
#         res *= i
#     return res


# from functools import reduce
# from operator import mul

# def almost_double_factorial(n):
#     return reduce(mul, range(1, n + 1, 2), 1)

# def almost_double_factorial(n):
    # return reduce(mul, [i for i in range(n+1) if i%2], 1)
```

Возвратите массив $B[0,\ldots, N-1]$, где $B_i = A_0 + \ldots + A_{i}$ --- массив частичных сумм массива $A$, предварительно удалив из массива $B$ все элементы, равные erase.
```python
def cumsum_and_erase(A, erase=1):
    B = [sum(A[:i]) for i in range(1, len(A) + 1) if sum(A[:i]) != erase]
    return B

A = [5, 1, 4, 5, 14]
B = cumsum_and_erase(A, erase=10)

assert B == [5, 6, 15, 29], "Something is wrong! Please try again"

# def cumsum_and_erase(m, erase=1):
#     s, res = 0, []   
#     for i in range(len(m)):
#         s+=m[i]
#         if s!=erase: res.append(s)
#     return res

# from itertools import accumulate

# def cumsum_and_erase(a, erase=1):
#     return list(filter((erase).__ne__, accumulate(a)))

# def cumsum_and_erase(A, erase=1):
    # return [sum(A[:i+1]) for i in range(len(A)) if sum(A[:i+1]) != erase]

# def cumsum_and_erase(A, erase = 1):
    # B = list(filter(lambda x: x != erase, map(lambda x: sum(A[:x+1]), range(len(A)))))
    # return B

# def cumsum_and_erase(A, erase = 1):
#     B = []
#     s = 0
#     for i in A:
#       s += i
#       B.append(s)
    
#     B = list(filter(lambda x : x != erase, B))
#     return B

# def cumsum_and_erase(a, erase = 1):
#     B = [sum(a[:i + 1]) for i in range(len(a))]
#     while erase in B:
#         B.remove(erase)
#     return B
```

Дан список текстов, слова в которых разделены пробелами (можно считать, что знаков препинания нет). Часть слов является "мусорными": в них присутствуют цифры и спецсимволы. Отфильтруйте такие слова из каждого текста. Используйте функции ``str.split``, ``str.isalpha``, ``str.join``, а также [генератор filter](http://pythonicway.com/python-functinal-programming).

```python
sentences = ['1 thousand devils', 'My name is 9Pasha', 'Room #125 costs $100']

def process(sentences):
    res = []
    for i in range(len(sentences)):
      a = sentences[i].split()
      b = list(filter(lambda x: x.isalpha(), a))
      a = ' '.join(b)
      res.append(a)
    
    result = res
    return result


assert process(sentences) == ['thousand devils', 'My name is', 'Room costs'], \
"Something is wrong! Please try again"

# def process(sentences):
    # return [" ".join(filter(lambda x: x.isalpha(), i.split())) for i in sentences]

# def process(sentences):
#     aa = [x.split() for x in sentences]
#     b = []
#     for x in aa:
#         bb = [y for y in x if y.isalpha()]
#         b.append(' '.join(bb))
#     return b

# def process(sentences):
#     result = []
#     for el in sentences:
#         words = el.split()
#         clean_words = [word for word in words if word.isalpha()]
#         result.append(' '.join(clean_words))
#     return result
```

### Собственный класс "Нейрон"
Реализуйте класс "Нейрон", у которого будет несколько методов: 

* ``__init__``. Принимает на вход массив весов нейрона --- $w = (w_1, \ldots, w_n)$, а также функцию активации $f$ (по умолчанию $f(x) = x$). Сохраняет веса и функцию внутри класса.

* ``forward``. Принимает на вход массив $x = (x_1, \ldots, x_N)$ --- входы нейрона. Возвращает $f(w_1x_1 + \ldots + w_nx_n)$.

* ``backlog``. Возвращает последний массив $x$, который подавался функции ``forward`` на вход. 

Если у вас появились трудности, информация по созданию своих классов есть здесь: https://pythonworld.ru/osnovy/obektno-orientirovannoe-programmirovanie-obshhee-predstavlenie.html

```python
class Neuron:
    
    def __init__(self, w, f = lambda x: x):
        self.w = w
        self.f = f
        self.x = list()
    
    def forward(self, x):
        self.x = x
        p = sum([x1 * y1 for x1, y1 in zip(x,w)])
        return self.f(p)
        
    def backlog(self):
        return self.x

# class Neuron:

#     def __init__(self, w, f = lambda x: x):
#         self.w = w
#         self.f = f

#     def forward(self, x):
#         self.x = x
#         return self.f(sum(list(map(lambda x, y: x * y, self.x, self.w))))

#     def backlog(self):
#         return self.x

# import numpy as np

# class Neuron:

#     def __init__(self, w, f = lambda x: x):
#         self.w = w
#         self.f = f
#         self.x = None

#     def forward(self, x):
#         self.x = x
#         return self.f(np.dot(self.w, self.x))

#     def backlog(self):
#         return self.x

# import numpy as np
# class Neuron:
#     def __init__(self, w, f = lambda x: x):
#         self.w = np.array(w)
#         self.act = f
    
#     def forward(self, x):
#         self.last_x = x
#         return self.act(self.w @ x)
        
#     def backlog(self):
#         return self.last_x
          
    # def backlog(self):
    #     try:
    #         return self.x
    #     except AttributeError :
    #         return None

    # def backlog(self):
    #     return self.x if hasattr(self, 'x') else None
```
