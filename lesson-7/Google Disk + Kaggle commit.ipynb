```python
hist = [{'order':456, 'item':2312, 'gmv':123, 'delivery':500},
        {'order':456, 'item':2313, 'gmv':456, 'delivery':500},
        {'order':456, 'item':2314, 'gmv':789, 'delivery':500}
        ]
```

```python
def allocate(arr):
    tot_cost = 0
    for cost in arr:
        tot_cost += cost['gmv']
    for item in arr:
        item['item_delivery_cost'] = round(item['delivery']*item['gmv']/tot_cost, 3)
    return arr
```

```python
allocate(hist)
```
