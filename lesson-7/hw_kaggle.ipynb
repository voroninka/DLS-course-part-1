```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
scoring='roc_auc'
parameters = [100, 10, 1, 0.1, 0.01, 0.001]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[feature_cols], data[target_col], 
                                                    train_size=0.8,
                                                    random_state=2)
```

Logistic Regression

```python
# Применяем секйлер
scaler = StandardScaler()

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])

# Применяем ванхот
ohe = OneHotEncoder(handle_unknown = 'ignore')

# Применяем ванхот к тесту
array_hot_encoded = ohe.fit_transform(X_train[cat_cols]).toarray()
feature_labels = ohe.get_feature_names_out(X_train[cat_cols].columns)
data_hot_encoded = pd.DataFrame(array_hot_encoded, index=X_train.index, columns=feature_labels)
data_other_cols = X_train.drop(columns=cat_cols)
X_train = pd.concat([data_hot_encoded, data_other_cols], axis=1)

# Применяем ванхот к трейну
array_hot_encoded = ohe.transform(X_test[cat_cols]).toarray()
feature_labels = ohe.get_feature_names_out(X_test[cat_cols].columns)
data_hot_encoded = pd.DataFrame(array_hot_encoded, index=X_test.index, columns=feature_labels)
data_other_cols = X_test.drop(columns=cat_cols)
X_test = pd.concat([data_hot_encoded, data_other_cols], axis=1)

# Логистическая регрессия
LogregCV_TTS = LogisticRegressionCV(Cs=parameters,cv=5,n_jobs=-1,scoring=scoring,refit=True)
LogregCV_TTS.fit(X_train,y_train)

print(f'train score: {LogregCV_TTS.score(X_train,y_train)}')
print(f'test score: {LogregCV_TTS.score(X_test,y_test)}')
print(f'Best C: {round(LogregCV_TTS.C_[0],1)}')
```

Logistic Regerssion + GridSearch

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder

data_pipe = data.copy()

ct = ColumnTransformer([
    ("scale",StandardScaler(),num_cols),
    ("oe",OrdinalEncoder(),cat_cols),
    ("ohe",OneHotEncoder(sparse=False),cat_cols),
],remainder='passthrough')

Pipe = make_pipeline(
    ct,
    LogisticRegression()
)

CV_pipe = GridSearchCV(estimator=Pipe, 
                        param_grid={'logisticregression__C':parameters},
                        cv=5, 
                        scoring=scoring,
                        n_jobs=-1)
CV_pipe.fit(data_pipe.loc[:,data_pipe.columns != target_col],data_pipe[target_col])

print(f'Score: {CV_pipe.best_score_}')
print(f'Best C: {list(CV_pipe.best_params_.values())[0]}')
```

Рандомсерч по параметрам с трейн-тест сплитом

```python
CBC = catboost.CatBoostClassifier(
                                  logging_level='Silent',
                                  cat_features=cat_cols,
                                  eval_metric='AUC:hints=skip_train~false',
                                  early_stopping_rounds=100
                                  )
grid = {
        'n_estimators': [100,200,500,1000,1200,1500],
        'learning_rate': [0.01, 0.03, 0.1],
        'depth': [3, 4, 5, 6],
        'l2_leaf_reg': np.linspace(0, 1, 10),
        'leaf_estimation_iterations': [5,10],
        }
CBC.randomized_search(grid, X=X_train_origin, y=y_train_origin, cv=5, n_iter=60, verbose=False);
```

```python
print('Train ROC AUC =', roc_auc_score(y_train_origin, CBC.predict_proba(X_train_origin)[:, 1]))
print('Test ROC AUC =', roc_auc_score(y_test_origin, CBC.predict_proba(X_test_origin)[:, 1]))
```

В Optuna задается функция objective, в которой прописываются параметры для оптимизации. Она подается в метод study для инициации обучения.

```python
def objective(trial):
    train_x, valid_x, train_y, valid_y = train_test_split(X_origin,y_origin, test_size=0.2, random_state = 2)

    param = {
        'iterations': trial.suggest_int("iterations", 100, 1200),
        'learning_rate': trial.suggest_float("learning_rate", 1e-3, 1e-1, log=True),
        "objective": trial.suggest_categorical("objective", ["Logloss", "CrossEntropy"]),
        "colsample_bylevel": trial.suggest_float("colsample_bylevel", 0.01, 0.1),
        "depth": trial.suggest_int("depth", 4, 12),
        'l2_leaf_reg': trial.suggest_float("l2_leaf_reg", 1e-8, 100.0, log=True),
        "boosting_type": trial.suggest_categorical("boosting_type", ["Ordered", "Plain"]),
        "bootstrap_type": trial.suggest_categorical(
            "bootstrap_type", ["Bayesian", "Bernoulli", "MVS"]
        ),
        'verbose': False,
    }

    if param["bootstrap_type"] == "Bayesian":
        param["bagging_temperature"] = trial.suggest_float("bagging_temperature", 0, 10)
    elif param["bootstrap_type"] == "Bernoulli":
        param["subsample"] = trial.suggest_float("subsample", 0.1, 1)

    gbm = catboost.CatBoostClassifier(cat_features=cat_cols,**param)

    gbm.fit(train_x, train_y, eval_set=[(valid_x, valid_y)], verbose=0, early_stopping_rounds=100)

    preds = gbm.predict_proba(valid_x)[:, 1]
    pred_labels = np.rint(preds)
    auc = roc_auc_score(valid_y, pred_labels)
    return auc
```

```python
optuna.logging.set_verbosity(optuna.logging.WARNING)
study = optuna.create_study(study_name="catboost", direction="maximize")
study.optimize(objective, n_trials=100, timeout=1200)
```

```python
trial = study.best_trial
for item, value in trial.params.items():
    print(f"{item}: {value}")
```
